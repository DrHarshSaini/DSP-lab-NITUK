Write a MATLAB program to perform the following basic operations for the
sequences:

 x(n) = {2,1,-4,3,2,4,-3} and y(n) = {1,4,2,5}.


Defining Custom Function for Padding and Operating it

```python
def padding(x, y, ox, oy):
    xsize = len(x)
    ysize = len(y)

    while ox < oy:
        x = np.insert(x, 0, 0)
        ox += 1
        xsize += 1

    while ox > oy:
        y = np.insert(y, 0, 0)
        oy += 1
        ysize += 1

    while xsize < ysize:
        x = np.append(x, 0)
        xsize += 1

    while xsize > ysize:
        y = np.append(y, 0)
        ysize += 1

    return x, y, ox

x = np.array([2, 1, -4, 3, 2, 4, -3])
y = np.array([1, 4, 2, 5])
ox = 3
oy = 1
x, y, oz = padding(x, y, ox, oy)
print("After Padding")
print("x[n]=",x)
print("y[n]=",y)
```

Q3. Write a MATLAB/python program to plot the following sequences for

x(n)= {1,2,3,4,6,8,5,4,3,2,1}. Origin is at 6

x1(n) = 3x(n-2) – 4x(n+3) 

x2(n) = 2x(2-n) + x(n)*x(n-3)

```python
x = np.array([1, 2, 3, 4, 6, 8, 5, 4, 3, 2, 1])
ox=6
n = np.arange(-(ox-1), len(x)-ox+1)

x_l2 = np.concatenate((np.zeros(2), x[:-2]))
x_r3 = np.concatenate((x[3:], np.zeros(3)))

x1 = 3 * x_l2 - 4 * x_r3

x_rev = np.concatenate((x[-2:], x[:-2]))
x_l3 = np.concatenate((np.zeros(3), x[:-3]))

x2 = 2 * x_rev + x * x_l3

plt.figure(figsize=(10, 8))

plt.subplot(3, 1, 1)
plt.stem(n, x)
plt.title('x(n)')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.stem(n, x1)
plt.title('x1(n)')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.stem(n, x2)
plt.title('x2(n)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

Q5). Write a MATLAB program to generate the following data s[n] and the noise d[n] and
perform ensemble average.

s[n] = 2*n((0.9)^n)

 for 0 ≤ n ≤ 71

Assume there are 50 measurements. Calculate MSE and relative error after
ensemble averaging. Use randn() function to generate and add noise.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
N = 72  # number of Samples
M = 50  # Number of Measurements
n = np.arange(N)  # Sample Indices

# generate the Signal s[n]
s = 2 * (0.9 ** n)

# Initialize arrays for Noise and ensemble average
noise = np.zeros((M, N))
ensemble_average = np.zeros(N)

# Create a Figure for plotting
plt.figure(figsize=(14, 8))

# gnerat noise, add it to the Signal, and perform ensemble averaging
for i in range(M):
    # Generate noise d[n] and add to signal s[n]
    d = np.random.randn(N)  # Noise
    noisy_signal = s + d    # Signal with noise
    
    # Store noise in the matrix
    noise[i, :] = d
    
    # Plot the noisy signal using stem plot
plt.subplot(3, 1, 1)
plt.stem(n, noisy_signal, linefmt='b-', markerfmt='bo', basefmt='b-')
plt.title('A Noisy Signal at Each Step')
plt.xlabel('n')
plt.ylabel('Amplitude')

# Compute ensemble average
ensemble_average = np.mean(noise + s, axis=0)

# Plot the original signal and the ensemble average
plt.subplot(3, 1, 2)
plt.stem(n, s, linefmt='g-', markerfmt='go', basefmt='g-', label='Original Signal')
plt.stem(n, ensemble_average, linefmt='r-', markerfmt='ro', basefmt='r-',label='Ensemble Average')
plt.title('Original Signal and Ensemble Average')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.legend()

# Plot a sample of noise
sample_noise = noise[0, :]
plt.subplot(3, 1, 3)
plt.stem(n, sample_noise, linefmt='k-', markerfmt='ko', basefmt='k-',label='Sample Noise')
plt.title('Sample Noise')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.legend()

# Adjust layout and show plot
plt.tight_layout()
plt.show()

# Compute Mean Squared Error (MSE)
MSE = np.mean((s - ensemble_average) ** 2)

# Compute relative error
relative_error = MSE / np.mean(s ** 2)

# Display results
print(f'Mean Squared Error (MSE): {MSE:.4f}')
print(f'Relative Error: {relative_error:.4f}')
```
